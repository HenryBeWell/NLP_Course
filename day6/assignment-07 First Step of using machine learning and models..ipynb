### Step2: 数据预处理

将pandas中的数据，依据是否是新华社的文章，请改变成新的数据dataframe: <content, y>, 其中，content是文章内容，y是0或者1. 你可能要使用到pandas的dataframe操作。https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.apply.html

```python
# 删除content为空的行
content.dropna(axis=0,subset=['content'],inplace=True)
content[:101]
```

```python
content['content'].isnull().value_counts()
```

```python
def judgement(new_df):
    if '新华' in new_df['content']: return 1
    else:
        return 0
new_df = pd.DataFrame(content,columns=['content','is_xinhua'])
new_df['is_xinhua'] = new_df.apply(judgement,axis=1)
new_df
```

```python
import jieba
import re
```

```python
new_content=[]
for i in range(len(new_df['content'])):
    tem = re.sub('[\\a-zA-Z0-9，。（）/：…@！？\s\n]', '', str(new_df.iloc[i, 0]))
    new_content.append(' '.join(jieba.cut(tem)))
new_content = pd.Series(new_content)
```

```python
new_content[0]
```

### Step3: 使用tfidf进行文本向量化

参考 https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.TfidfVectorizer.html 对文本进行向量化

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
```

```python
vectorrizer = TfidfVectorizer()
X_tfidf = vectorrizer.fit_transform(new_content)
```
